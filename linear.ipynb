```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

 Step 2: Data Preparation
 -

```python
# Generate synthetic data for the example
np.random.seed(0)
hours_studied = np.random.uniform(0, 10, 100)
```

```python
exam_scores = 50 + 5 * hours_studied + np.random.normal(0, 2, 100)  # Linear relationship with noise
```

```python
# Create a DataFrame
data = pd.DataFrame({'Hours_Studied': hours_studied, 'Exam_Scores': exam_scores})
data
```

```python

# Plot the data
plt.scatter(data['Hours_Studied'], data['Exam_Scores'])
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.title('Hours Studied vs Exam Scores')
plt.show()
```

Step 3: Model Training
-

```python
# Split the data into training and testing sets
X = data[['Hours_Studied']]
y = data['Exam_Scores']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Initialize and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

Step 4: Model Evaluation
-

```python
# Predict using the test set
y_pred = model.predict(X_test)

# Calculate performance metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")

# Plot the regression line
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, y_pred, color='red', linewidth=2)
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.title('Regression Line')
plt.show()
```
